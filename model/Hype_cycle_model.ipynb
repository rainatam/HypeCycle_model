```python
import json
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Loading data

```python
# time span
start = 1960
end = 2016
time_span = [i for i in range(start, end)]

# terminology
with open('../data/term_list.json', 'r') as f:
    term_data = json.load(f)
    
terms = term_data['L0'] + term_data['L1'] + term_data['L2'] + term_data['Test']
terms = list(set(terms))

# papers
with open('../data/paper_dict.json', 'r') as f:
    paper_dict = json.load(f)

total_papers = {} 
with open('../data/total_paper_dict.json', 'r') as f:
    data = json.load(f)
    for year in time_span:
        total_papers[year] = data[str(year)]

# patents        
with open('../data/patent_dict.json', 'r') as f:
    patent_dict = json.load(f)

total_patents = {} 
with open('../data/total_patent_dict.json', 'r') as f:
    data = json.load(f)
    for year in time_span:
        total_patents[year] = data[str(year)]
    
print('#terms =', len(terms))
```

```python
term_list = []
paper_list = []
patent_list = []

def cumulate(series):
    cum = 0
    cum_series = []
    for i in range(len(series)):
        cum += series[i]
        cum_series.append(cum)
    return np.array(cum_series)

for term in terms:
    paper = []
    for year in time_span:
        paper.append(paper_dict[term][str(year)])
    paper = np.array(paper, dtype=np.float64)
            
    patent = []
    for year in time_span:
        patent.append(patent_dict[term][str(year)] if year >= 1976 else 0)
    patent = np.array(patent, dtype=np.float64)

    if np.sum(paper > 0) > 5 and np.sum(paper) >= 500:
        paper_list.append(paper)
        patent_list.append(patent)
        term_list.append(term)
        
N = len(paper_list)
assert N == len(patent_list)
assert N == len(term_list)
print('#terms =', N)
```

# Normalization

```python
paper_norm_list = []
patent_norm_list = []
patent_norm_cum_list = []
for i in range(N):
    x = paper_list[i].copy()
    y = patent_list[i].copy()
    
    for j in range(len(time_span)): x[j] /= total_papers[time_span[j]]
    for j in range(len(time_span)): y[j] /= total_patents[time_span[j]] if total_patents[time_span[j]] > 0 else 1
    z = cumulate(y)
    
    paper_norm_list.append(x)
    patent_norm_list.append(y)
    patent_norm_cum_list.append(z)
```

# Labeling Status

```python
from sklearn.linear_model import LogisticRegression, LinearRegression

# 3 status of trend
INC = 0
DEC = 1
FLAT = 2

L = 5

status_color = ['c', 'black', 'gray']

def cal_slope(series):
    yi = series.reshape(-1, 1)
    xi = np.array([i for i in range(len(series))]).reshape(-1, 1)

    model = LinearRegression()
    model.fit(xi, yi)
    k = model.coef_[0][0]

    if k > 0.01:
        k = 1.0
    elif k < -0.01:
        k = -1.0
    else:
        k = 0.0
        
    return k

def cal_status(value):
    if value > 0.3:
        return INC
    elif value < -0.3:
        return DEC
    else:
        return FLAT
    
def smooth(series):
    for i in range(len(series)):
        if i > 0 and i < len(series)-1 and series[i] != series[i-1] and series[i-1] == series[i+1]:
            series[i] = series[i-1]
        elif i > 1 and i < len(series)-3 and series[i] == series[i+1] and series[i] != series[i-1] and \
             series[i-1] == series[i+2] and series[i-2] == series[i-1] and series[i+2] == series[i+3]:
            series[i] = series[i-1]
            series[i+1] = series[i-1]
    return series

def label_status(x, d=0.3):
    labels = []
    slopes = []
    for i in range(len(x)):
        if i == 0:
            slope = 0
        elif i == len(x) - 1:
            slope = slopes[-1]
        elif i - L <= 0:
            slope = cal_slope(x[0:i])
        elif i + L >= len(x):
            pred = cal_slope(x[i:])
        else:
            slope = cal_slope(x[i-L : i+L]) if np.sum(x[i-L : i+L] > 0) > 0 else 0
            
        # weighted smoothing
        mean_slope = slope * d + slopes[-1] * (1-d) if i > 0 else slope
        slopes.append(mean_slope)
        labels.append(cal_status(mean_slope))
    labels = smooth(labels)
    return labels
```

# Time series segmentation

```python
def calculate_slope(series):
    yi = series.reshape(-1, 1)
    xi = np.array([i for i in range(len(series))]).reshape(-1, 1)

    model = LinearRegression()
    model.fit(xi, yi)
    k = model.coef_[0][0]
    return k

def approximate_line(series):
    yi = series.reshape(-1, 1)
    xi = np.array([i for i in range(len(series))]).reshape(-1, 1)

    model = LinearRegression()
    model.fit(xi, yi)
    slope = model.coef_[0][0]
    line = model.predict(xi).reshape(-1)
    return line, slope

def calculate_error(seq_i, seq_j, lbl_i, lbl_j):
    slope_i = calculate_slope(seq_i)
    slope_j = calculate_slope(seq_j)
    error = abs(slope_i - slope_j)
    if np.all(lbl_i[-int(len(lbl_i)*0.8):] == INC) and np.all(lbl_j[:int(len(lbl_j)*0.8)] == INC): error = 0
    if np.all(lbl_i[-int(len(lbl_i)*0.8):] == DEC) and np.all(lbl_j[:int(len(lbl_j)*0.8)] == DEC): error = 0
    return error

def bottom_up(series, labels, max_error, step=5, k=2):
    seg_list = []
    idx_list = []
    lbl_list = []
    for i in range(0, len(series)-1, step):
        if i+step >= len(series)-1: step = len(series)-i
        seg_list.append(series[i:i+step])
        lbl_list.append(np.array(labels[i:i+step]))
        idx_list.append((i, i+step))

    merge_cost = []
    for i in range(len(seg_list)-1):
        merge_cost.append(calculate_error(seg_list[i], seg_list[i+1], lbl_list[i], lbl_list[i+1]))

    while np.min(merge_cost) < max_error:
        i = np.argmin(merge_cost)
        seg_list[i] = np.hstack((seg_list[i], seg_list[i+1]))
        lbl_list[i] = np.hstack((lbl_list[i], lbl_list[i+1]))
        idx_list[i] = (idx_list[i][0], idx_list[i+1][1])
        del(seg_list[i+1])
        del(lbl_list[i+1])
        del(idx_list[i+1])
        del(merge_cost[i])
        if(len(seg_list)==k): break
        if i < len(seg_list)-1: merge_cost[i] = calculate_error(seg_list[i], seg_list[i+1], lbl_list[i], lbl_list[i+1])
        merge_cost[i-1] = calculate_error(seg_list[i-1], seg_list[i], lbl_list[i-1], lbl_list[i])
    return seg_list, idx_list
    
def combine_segment(x, label_list, seg_list, idx_list,):
    slope_list = []

    for seg, idx in zip(seg_list, idx_list):
        line, slope = approximate_line(seg)
        slope_list.append(slope)

    segments = []
    
    # upward segments
    slope_threshold = 0.005
    delta_threshold = 0.3
    segment = [0, 0]
    for idx, slope in zip(idx_list, slope_list):
        if slope > slope_threshold:
            segment[1] = idx[1]
        else:
            if segment[0] < segment[1] and np.max(x[segment[1]-3:segment[1]]) - np.min(x[segment[0]:segment[0]+3]) > delta_threshold:
                segments.append([segment[0], segment[1], INC])
            segment = [idx[1], idx[1]]
    if segment[0] < segment[1] and np.max(x[segment[1]-3:segment[1]]) - np.min(x[segment[0]:segment[0]+3]) > delta_threshold:
        segments.append([segment[0], segment[1], INC])
        
    # downward segments
    segment = [0, 0]
    for idx, slope in zip(idx_list, slope_list):
        if slope < -slope_threshold:
            segment[1] = idx[1]
        else:
            if segment[0] < segment[1]:
                segments.append([segment[0], segment[1], DEC])
            segment = [idx[1], idx[1]]
    if segment[0] < segment[1]:
        segments.append([segment[0], segment[1], DEC])
    
    segments.sort() 
    return segments
    
def extract_bell_curve(x, label_list, segments):
    bell_curves = []
    
    def append(st, ed):
        while st-1 >= 0 and label_list[st-1] == INC: 
            st -= 1
        while ed+1 <= len(x) and label_list[ed] == DEC: 
            ed += 1
        bell_curves.append((st, ed))
    
    
    if len(segments) == 1:
        st, ed, _ = segments[0]
        append(st, ed)
    elif np.all(np.array([seg[2] for seg in segments]) == INC) or np.all(np.array([seg[2] for seg in segments]) == DEC):
        for seg in segments:
            if np.any(x[seg[0]:seg[1]] > 0.8):
                st, ed, _ = seg
                append(st, ed)                                          
    else:
        if segments[0][2] == DEC and np.any(x[segments[0][0]:segments[0][1]] > 0.7):
            st, ed, _ = segments[0]
            append(st, ed)
        for i in range(len(segments)-1):
            if segments[i][2] == INC and segments[i+1][2] == DEC and np.any(x[segments[i][0]:segments[i+1][1]] > 0.7) and np.max(x[segments[i][0]:segments[i+1][1]]) - np.min(x[segments[i][0]:segments[i+1][1]]) > 0.5:
                st, ed = segments[i][0], segments[i+1][1]
                append(st, ed)
            elif segments[i][2] == INC and segments[i+1][2] == INC and np.any(x[segments[i][0]:segments[i+1][0]] > 0.7) and np.max(x[segments[i][0]:segments[i+1][0]]) - np.min(x[segments[i][0]:segments[i+1][0]]) > 0.5:
                st, ed = segments[i][0], segments[i+1][0]
                append(st, ed)
        if segments[-1][2] == INC and np.any(x[segments[-1][0]:segments[-1][1]] > 0.8):
            st, ed, _ = segments[-1]
            append(st, ed)
            
    dedup_bell_curves = []
    i = 0
    while i < len(bell_curves)-1:
        remove = True
        if bell_curves[i][1] < bell_curves[i+1][0] and np.any(x[bell_curves[i][1]:bell_curves[i+1][0]] < 0.3):
            remove = False
        elif bell_curves[i][1] >= bell_curves[i+1][0]:
            st = max(0, bell_curves[i+1][0]-3)
            ed = min(len(x), bell_curves[i][1]+3)
            if np.any(x[st:ed] < 0.3): remove = False

        if remove:
            if min(np.max(x[bell_curves[i][0]:bell_curves[i][1]]), np.max(x[bell_curves[i+1][0]:bell_curves[i+1][1]])) > 0.8:
                dedup_bell_curves.append((bell_curves[i][0], bell_curves[i+1][1]))
            elif np.max(x[bell_curves[i][0]:bell_curves[i][1]]) <= np.max(x[bell_curves[i+1][0]:bell_curves[i+1][1]]):
                dedup_bell_curves.append(bell_curves[i+1])
            else:
                dedup_bell_curves.append(bell_curves[i])
        else:
            dedup_bell_curves.append(bell_curves[i])
            dedup_bell_curves.append(bell_curves[i+1])
        i += 2
    if i < len(bell_curves): dedup_bell_curves.append(bell_curves[-1])
 
    return dedup_bell_curves
```

# Fitting curves
### Hype cycle = hype stage (bell curve) + implementation stage (S-curve)

```python
from scipy import interpolate
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

def gompertz(x, a, k, xc):
    return a * np.exp( -np.exp( -k * (x - xc) ) )

def cumulate(series):
    cum = 0
    cum_series = []
    for i in range(len(series)):
        cum += series[i]
        cum_series.append(cum)
    return np.array(cum_series)

def non_cumulate(series):
    noncum_series = [series[0]]
    for t in range(1, len(series)):
        noncum_series.append(series[t] - series[t-1])
    return np.array(noncum_series)

def plot_line(x_list, color):
    for x in x_list:
        plt.plot([x, x],[0, 1], ':', c=color)
```

```python
def simulate_hype_cycle(x, y, term, norm=False, plot=True):
    xy_max = max(np.max(x), np.max(y))
    if xy_max == 0: xy_max = 1
    if norm:
        x /= xy_max
        y /= xy_max
    
    x_max = np.max(x)
    x_norm = x / x_max if np.max(x) > 0 else x
    y_max = np.max(y) 
    y_norm = y / y_max if np.max(y) > 0 else y
    height = max(x_max, y_max)
    labels = label_status(x_norm)

    seg_list, idx_list = bottom_up(x_norm, labels, 0.02, k=1, step=5)
    segments = combine_segment(x_norm, labels, seg_list, idx_list)
    bell_curves = extract_bell_curve(x_norm, labels, segments)

    hype_curves = []
    impl_curve = None
    # ================================= GRAPH ==================================
    if plot:
        plt.figure(figsize=(15, 4))
        plt.title("%d %s"%(i, term))
        if norm: plt.plot(range(len(x)), x_norm, ':', color='lightgray')
        if norm: plt.plot(range(len(y)), y_norm, ':', color='lightgray')
        plt.plot(range(len(x)), x, '-', label='papers')
        plt.plot(range(len(y)), y, '-', label='patents')

    # =============================== Hype stage ===============================
    colors = [['blue', 'navy'], ['green', 'darkgreen'], ['orange', 'darkorange']]
    for bell_curve, c in zip(bell_curves, colors):
        st, ed = bell_curve
        if st < ed:
            scale = np.max(x_norm[st:ed])
            if plot: plt.plot([st, st], [0, height], '--', color=c[0])
            if plot: plt.plot([ed-1, ed-1], [0, height], '--', color=c[0])
            st = max(st - 0, 0)
            ed = min(ed + 0, len(x))

            hype_stage = x_norm[st:ed]
            reverse = np.argmax(hype_stage) + st < 3
            if reverse: hype_stage = hype_stage[::-1]
            cum_hype_stage = cumulate(hype_stage)
            ydata = cum_hype_stage
            xdata = range(len(ydata))
            popt, pcov = curve_fit(gompertz, xdata, ydata, maxfev=500000)
            xdata = range(-st, len(x)-st) if not reverse else range(ed-len(x), len(ydata))
            hype_curve = non_cumulate(gompertz(xdata, *popt))
            hype_curve /= np.max(hype_curve)
            if hype_curve[0] - hype_curve[1] > 0.1:
                hype_curve[0] = hype_curve[1] - abs(hype_curve[2]-hype_curve[1])
                hype_curve /= np.max(hype_curve)
            hype_curve *= scale
            if reverse: hype_curve = hype_curve[::-1]
            hype_curve *= x_max

            if plot: plt.plot(range(len(x)), hype_curve, '--', linewidth=3, color=c[1], label='hype stage')
            hype_curves.append(hype_curve)
    # =============================== Impl stage ===============================
    ydata = y_norm[8:]
    xdata = range(len(ydata))
    popt, pcov = curve_fit(gompertz, xdata, ydata, maxfev=500000)
    xdata = range(-8, len(ydata))
    impl_curve = gompertz(xdata, *popt)
    impl_curve /= np.max(impl_curve)
    if impl_curve[0] >= 0.05:
        ydata = y
        xdata = range(len(ydata))
        popt, pcov = curve_fit(gompertz, xdata, ydata, maxfev=500000)
        impl_curve = gompertz(xdata, *popt)
        impl_curve /= np.max(impl_curve)
    impl_curve *= y_max
    if plot: plt.plot(range(len(y)), impl_curve, '--', linewidth=3, color='darkred', label='impl. stage')
    if plot: plt.legend(loc='upper right')
        
    return hype_curves, impl_curve
```

# Dividing four periods of hype cycle

```python
def select_over_threshold(x, threshold, offset=0):
    st, ed = None, None
    x = x / np.max(x)
    peak = np.argmax(x)
    if peak < threshold:
        return st, ed
    st, ed = peak, peak
    while (st-1 >= 0 and x[st-1] >= threshold) or (st-2 >= 0 and x[st-2] >= threshold) or (st-3 >= 0 and x[st-3] >= threshold): 
        st -= 1
    while (ed+1 < len(x) and x[ed+1] >= threshold) or (ed+2 < len(x) and x[ed+2] >= threshold) or (ed+3 < len(x) and x[ed+3] >= threshold):
        ed += 1
    if ed == st: return None, None
    return st+offset, ed+offset

def get_periods(x, y, hype_cycle, peaks, nadir, threshold):
    n = len(hype_cycle)
    periods = [[], [], [], []]
    if len(peaks) > 0:
        hypes = []
        st, ed = 0, 0
        for i in range(len(peaks)-1):
            st = ed
            ed = -np.argmin(hype_cycle[peaks[i+1]:peaks[i]:-1]) + peaks[i+1]
            hypes.append((st, ed))
        hypes.append((ed, nadir if nadir > 0 else n-1)) 

        slope = False
        for hype, peak in zip(hypes, peaks):
            hype_st, hype_ed = hype
            peak_st, peak_ed = select_over_threshold(x[hype_st:hype_ed+1], 0.8, hype_st)
            if peak == n-1 and np.argmax(x) == n-1: 
                if hype_st < n-1: periods[0].append((hype_st, hype_ed+1))
            elif peak_st is None or peak_ed is None:
                if peak == 0:
                    periods[2].append((hype_st, hype_ed+1))
                    slope = True
                else:
                    if hype_st < n-1: periods[0].append((hype_st, hype_ed+1))
            else:
                slope = True
                if hype_st < n-1: periods[0].append((hype_st, peak_st+1))
                if peak_st < n-1: periods[1].append((peak_st, peak_ed+1))
                if peak_ed < n-1: periods[2].append((peak_ed, hype_ed+1))

        if nadir > 0 and slope:
            for t in range(nadir, n):
                if hype_cycle[t] > threshold or t == n-1:
                    tmp = periods[2][-1][0]
                    periods[2][-1] = (tmp, t+1)
                    hype_ed = t
                    break

        if hype_ed < n-1: periods[3].append((hype_ed, n))
    else:
        for t in range(n):
            if hype_cycle[t] > threshold:
                periods[0].append((0, t+1))
                if t < n-1: periods[3].append((t, n))
                break

    return periods

def plot_divided_hype_cycle(hype_cycle, periods, offset=0):
    name = ['Rise', 'Peak', 'Trough', 'Plateau']
    color = ['deepskyblue', 'lawngreen', 'coral', 'mediumpurple']

    for i in range(4):
        if len(periods[i]) > 0:
            st, ed = periods[i][0]
            plt.plot(range(offset+st, offset+ed), hype_cycle[st:ed], '-', linewidth = 3, c=color[i], label=name[i])
            for period in periods[i][1:]:
                st, ed = period
                plt.plot(range(offset+st, offset+ed), hype_cycle[st:ed], '-', linewidth = 3, c=color[i])
                
def divide_hype_cycle(x, y, term, hype_stages, impl_stage, plot=True):
    # =============================== Peaks ===============================
    xy_max = max(np.max(x), np.max(y))
    if xy_max == 0.: xy_max = 1.
    x /= xy_max
    y /= xy_max
    hype_stages = [ h / xy_max for h in hype_stages ]
    impl_stage /= xy_max
    
    impl_st = 0
    hype_stage_max = 0
    peaks = []
    if len(hype_stages) > 0:
        hype_stage_max = np.max([np.max(hype_stage) for hype_stage in hype_stages])
        for hype_stage in hype_stages:
            peak = np.argmax(hype_stage)
            if hype_stage[peak] > impl_stage[peak]:
                impl_st = max(impl_st, peak)
                peaks.append(peak)
        peaks.sort()
    
    # ========================= Combine two stages =========================
        hype_cycle = []
        for j in range(len(x)):
            if np.isnan(impl_stage[j]) or np.isinf(impl_stage[j]): impl_stage[j] = 0
            if j <= impl_st:
                hype_cycle.append(max([hype_stage[j] for hype_stage in hype_stages]))
            else:
                hype_cycle.append(max([hype_stage[j] for hype_stage in hype_stages] + [impl_stage[j]]))
    else:
        hype_cycle = impl_stage
    # =============================== Nadir ===============================
    nadir = np.argmin(hype_cycle[peaks[-1]:]) + peaks[-1] if len(peaks) > 0 else 0
    if nadir == len(hype_cycle)-1: nadir = 0
        
    st, ed = 0, 0
    hypes = []
    if len(peaks) > 0:
        for j in range(len(peaks)-1):
            st = ed
            ed = -np.argmin(hype_cycle[peaks[j+1]:peaks[j]:-1]) + peaks[j+1]
            hypes.append((st, ed))
    
    periods = get_periods(x, y, hype_cycle, peaks, nadir, max(0.2, hype_cycle[nadir] * 1.2))

    # ================================= Graph =================================
    if plot:
        plt.figure(figsize=(15, 4))
        xy_max = max(np.max(x), np.max(y))
        x /= xy_max
        y /= xy_max
        plt.plot(range(len(x)), x / np.max(x), ':', c='lightgray')
        plt.plot(range(len(y)), y / np.max(y), ':', c='lightgray')
        plt.plot(range(len(x)), x)
        plt.plot(range(len(y)), y)
        
        plt.plot(range(len(time_span)), hype_cycle, '-', linewidth = 2, c='yellow')
        for hype_stage in hype_stages:
            plt.plot(range(len(time_span)), hype_stage, '--', linewidth = 2, c='navy')
        plt.plot(range(len(time_span)), impl_stage, '--', linewidth = 2, c='darkred')
    
        if len(hype_stages) > 0:
            for peak in peaks: plot_line([peak+0], 'coral')
            for hype in hypes: plot_line([hype[1]+0], 'mediumpurple')
            if nadir > 0: plot_line([nadir+0], 'mediumpurple')
                
        plot_divided_hype_cycle(hype_cycle, periods, offset=0)
        plt.legend(loc='best')
        plt.title("%d %s"%(i, term))
        
    return hype_cycle, periods
```
